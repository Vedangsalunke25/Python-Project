```python
def get_subjects(num_subjects):
    subjects = []
    teachers = []
    for i in range(num_subjects):
        subject = input(f"Enter subject {i + 1}: ")
        subjects.append(subject)
        teacher = input(f"Enter teacher for {subject}: ")
        teachers.append(teacher)
    return subjects, teachers


def generate_timetable(num_subjects, max_theory_lectures):
    subjects, teachers = get_subjects(num_subjects)
    timetable = {"Monday": [], "Tuesday": [], "Wednesday": [], "Thursday": [], "Friday": []}
    subject_max_theory_lectures = max_theory_lectures // num_subjects

    subject_index = 0
    for day in timetable:
        scheduled_lectures = []
        theory_counter = 0
        practical_counter = 0

        while theory_counter < subject_max_theory_lectures :
            subject = subjects[subject_index]
            teacher = teachers[subject_index]

            if subject not in scheduled_lectures:

                    timetable[day].append(subject + " - " + teacher)
                    scheduled_lectures.append(subject)
                    theory_counter += 1

            subject_index = (subject_index + 1) % num_subjects

    return timetable


def print_timetable(timetable):
    print("Timetable:")
    for day, lectures in timetable.items():
        print(day + ":")
        for lecture in lectures:
            print(lecture)
        print()


num_subjects = int(input("Enter the number of subjects: "))
max_theory_lectures = int(input("Enter the maximum number of theory lectures: "))

timetable = generate_timetable(num_subjects, max_theory_lectures)
print_timetable(timetable)
```
